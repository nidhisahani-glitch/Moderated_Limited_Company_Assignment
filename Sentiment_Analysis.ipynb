```python
import pandas as pd
import numpy as np
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
# Download the 'punkt_tab' data for sentence tokenization
nltk.download('punkt_tab') # This line is added to download the missing 'punkt_tab' data

# Load the dataset
file_path = '/content/sentimentdataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Original Data:")
print(data.head())


# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Alternatively, fill missing values with a default value (e.g., empty string)
# data.fillna('', inplace=True)

# Step 2: Standardize Text
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Assuming the text column is named 'text', apply the cleaning function
data['cleaned_text'] = data['Text'].apply(clean_text)


# Encode labels
if 'Sentiment' in data.columns:  # Check if there is a label column
    data['Sentiment'] = pd.factorize(data['Sentiment'])[0]  # Convert labels to numerical format

# Display the cleaned data
print("\nCleaned Data:")
print(data[['Text', 'cleaned_text', 'Sentiment']].head())

# Save the cleaned data to a new CSV file
data.to_csv('new_sentimentdataset.csv', index=False)  # Update with your desired path
```

```python
import pandas as pd
from datasets import Dataset

# Load your cleaned dataset
data = pd.read_csv('/content/new_sentimentdataset.csv')

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(data[['cleaned_text', 'Sentiment']])

# Split the dataset into train and test sets
train_test_split = dataset.train_test_split(test_size=0.2)
train_dataset = train_test_split['train']
test_dataset = train_test_split['test']
```

```python
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification

# Load a model suitable for sequence classification
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert/distilbert-base-uncased", num_labels=len(set(data['Sentiment']))
)
# Tokenize the dataset and include labels
def tokenize_function(examples):
    # Tokenize the text
    tokenized_inputs = tokenizer(
        examples['cleaned_text'], padding="max_length", truncation=True
    )
    # Include labels in the output
    tokenized_inputs['labels'] = examples['Sentiment']
    return tokenized_inputs

tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_test_dataset = test_dataset.map(tokenize_function, batched=True)

# Set training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    # Enable debugging with CUDA_LAUNCH_BLOCKING
    # This might give more specific error messages
    no_cuda=False,  # CUDA is enabled
    # Remove the following line after debugging:
    # fp16=False  # Disable mixed precision training for debugging
)

# Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_test_dataset,
)

# Train the model
trainer.train()
```

```python
# Evaluate the model
results = trainer.evaluate()
print(results)
```

```python
from transformers import pipeline

# Initialize sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Function to analyze sentiment and map to an emotion
def analyze_sentiment(input_text):
    result = sentiment_pipeline(input_text)
    sentiment = result[0]["label"].lower()  # Extract sentiment (e.g., "positive" or "negative")
    return sentiment

def map_emotion(sentiment):
    # Map sentiment to emotional states
    emotion_mapping = {
        "positive": "happy",
        "negative": "sad",
        "neutral": "neutral"
    }
    return emotion_mapping.get(sentiment, "neutral")

# Simulated chatbot response generator
def generate_chatbot_response(input_text, predicted_emotion, cultural_context):
    # Simple simulated response generator
    if predicted_emotion == "happy":
        return "That's wonderful! I'm so glad to hear that."
    elif predicted_emotion == "sad":
        return "I'm sorry to hear that. Is there something specific bothering you?"
    elif predicted_emotion == "neutral":
        return "Got it. Is there anything else you'd like to share?"
    else:
        return "I'm here to help. How can I assist you?"

# Function to evaluate chatbot performance
def evaluate_chatbot(test_data):
    emotion_matches = []
    response_matches = []

    for test in test_data:
        # Analyze sentiment and map to emotion
        predicted_sentiment = analyze_sentiment(test["input"])
        predicted_emotion = map_emotion(predicted_sentiment)
        is_emotion_correct = predicted_emotion == test["expected_emotion"]
        emotion_matches.append(is_emotion_correct)

        # Generate chatbot response
        generated_response = generate_chatbot_response(
            test["input"], predicted_emotion, test["cultural_context"]
        )

        # Check response relevance
        is_response_relevant = all(
            word in generated_response.lower() for word in test["expected_response"].lower().split()
        )
        response_matches.append(is_response_relevant)

        # Log test case details
        print(f"User Input: {test['input']}")
        print(f"Expected Emotion: {test['expected_emotion']}, Predicted Emotion: {predicted_emotion}")
        print(f"Emotion Match: {'✔' if is_emotion_correct else '✘'}")
        print(f"Expected Response: {test['expected_response']}")
        print(f"Generated Response: {generated_response}")
        print(f"Response Match: {'✔' if is_response_relevant else '✘'}")
        print("-" * 50)

    # Summary results
    emotion_accuracy = sum(emotion_matches) / len(emotion_matches) * 100
    response_accuracy = sum(response_matches) / len(response_matches) * 100

    print("\n=== Summary Results ===")
    print(f"Emotion Detection Accuracy: {emotion_accuracy:.2f}%")
    print(f"Response Relevance Accuracy: {response_accuracy:.2f}%")

# Example test data
test_data = [
    {
        "input": "I'm feeling down and hopeless.",
        "expected_emotion": "sad",
        "expected_response": "I'm sorry to hear that. Is there something specific bothering you?",
        "cultural_context": "neutral",
    },
    {
        "input": "Life is amazing today!",
        "expected_emotion": "happy",
        "expected_response": "That's wonderful! I'm so glad to hear that.",
        "cultural_context": "neutral",
    },
    {
        "input": "I'm angry at how unfair things are!",
        "expected_emotion": "angry",
        "expected_response": "I'm sorry you're feeling this way. Would you like to talk about it?",
        "cultural_context": "neutral",
    },
]

# Evaluate chatbot performance
evaluate_chatbot(test_data)
```
